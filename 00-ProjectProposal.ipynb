## Research Question

What specific gameplay metrics (e.g., headshot percentage, utility damage per round, trade kill frequency, and first-duel success rate) best classify a CS2 player as "high-performing" (e.g., Premier Rating 20k+ / Faceit Level 8-10) versus "average" (Premier Rating 10k / Faceit Level 4-5)? 


## Background and Prior Work

Counter-Strike 2 (CS2) is a competitive tactical shooter where player performance is shaped by both mechanical skill and decisionmaking. At a basic level, players are often evaluated using visible statistics such as kills, deaths, assists, headshot percentage, and damage. However, these simple statistics do not always fully explain why one player performs better than another. For example, a player may not always have the most kills, but they may consistently win opening duels, trade teammates effectively, use utility to create advantages, or deal high damage in important rounds. Because of this, our project asks which specific gameplay metrics best distinguish “high performing” CS2 players, such as Premier Rating 20k+ or FACEIT Level 8–10 players, from more average players, such as Premier Rating 10k or FACEIT Level 4–5 players. FACEIT’s own documentation describes its skill levels as being directly linked to Elo ratings, which are intended to reflect a player’s perceived skill and game knowledge.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Previous work in Counter-Strike analytics has often focused on predicting team or round outcomes rather than directly classifying individual player skill tiers. For example, one commonly used CS:GO round winner dataset contains round snapshots from professional matches and includes variables such as round time, map, bomb plant status, player health, armor, money, equipment, and weapons.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Projects using this dataset usually frame the problem as a binary classification task: predicting whether the Terrorist or Counter-Terrorist side will win a round based on the current game state. This type of work is useful because it shows that machine learning can extract meaningful signals from Counter-Strike match data, but its main target is the outcome of a round, not the long term difference between high skill and average individual players.

A more directly analytical approach is found in Xenopoulos, Doraiswamy, and Silva’s paper “Valuing Player Actions in Counter-Strike: Global Offensive.” Their work introduced a data model and context-aware framework for estimating how individual player actions change a team’s probability of winning. Using over 70 million ingame CS:GO events, they showed that player actions can be evaluated beyond simple scoreboard statistics by measuring their contribution to win probability.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) This is closely related to our project because both studies are interested in understanding player contribution. However, their research focuses on action valuation and win probability modeling, while our project focuses on identifying which player level metrics are most useful for separating ranked skill groups.

Community and industry rating systems also show that Counter-Strike performance is multi-dimensional. HLTV’s Rating 2.0 was designed to summarize professional player performance using more than just kills. HLTV states that it incorporates factors such as damage dealt, opening kills, 1vX wins, traded deaths, and other statistics to better capture the different ways players contribute in a match.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) HLTV later introduced Rating 2.1 for Counter-Strike 2, adjusting the rating system for CS2 statistics and reducing the reward for saving in lost rounds.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) These rating systems support the idea that individual performance cannot be measured accurately with one simple statistic, but they are mainly designed as aggregate ratings for professional matches rather than as an interpretable classification model for ranked players.

Our project builds on these previous approaches by shifting the focus from predicting round winners or creating a single overall rating to identifying which specific metrics best separate skill tiers. Metrics such as headshot percentage, utility damage per round, trade kill frequency, opening-duel or first-duel success rate, average damage per round, and K/D ratio may each capture different aspects of performance. By comparing these metrics between high performing and average players, we hope to determine which features are most predictive of skill level and which features may be less useful than players commonly assume. This can help create a more interpretable understanding of CS2 performance and provide insight into what measurable behaviors are most associated with higher ranked play.

## Hypothesis


We hypothesize that mechanical metrics, such as headshot percentage and raw damage per round, will establish a baseline separating low tier players from average players. Tactical and team-oriented metrics, specifically utility damage per round and trade kill percentage will be the primary features distinguishing average players from high-performing players. We predict this outcome because mechanical aim often reaches a plateau at the upper-middle ranks, meaning players must rely on superior utility usage and coordinated team trading to secure an advantage.
